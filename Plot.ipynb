```python
import matplotlib.pyplot as plt
import math
```

```python
filepath = 'out.txt'
with open(filepath) as fp:
   line = fp.readline()
   
   u = []
   while line:
       line = fp.readline()
       if len(line) > 0:
           u.append(float(line.replace("\n", "")))
```

```python
x = []
y = []
a = 2*math.pi/ len(u)

for i in range(0,len(u)):
    x.append(2*math.cos(a*i))
    y.append(2*math.sin(a*i))
```

```python
sc = plt.scatter(x,y,c=u)
clb = plt.colorbar()
clb.ax.set_title('u')
```
