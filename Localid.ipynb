```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 2, 4
```

```python
def locid(i,np,nl):
    return int(i / (np/nl))
```

```python
for i in range(0,4):
    plt.scatter(locid(i,4,2),i)
plt.xlabel("Localitiy")
plt.ylabel("Partitions")
plt.xticks([0,1])
plt.yticks([0,1,2,3])
plt.title("np=4 and nl=2")
plt.savefig("partition1.pdf")
```

```python
for i in range(0,4):
    plt.scatter(locid(i,4,1),i)
plt.xlabel("Localitiy")
plt.ylabel("Partitions")
plt.xticks([0])
plt.yticks([0,1,2,3])
plt.title("np=4 and nl=1")
plt.savefig("partition2.pdf")
```
